```python
import matplotlib.pyplot as plt
import numpy as np
```

# types of graphs
1. plot-        line graph,between two parameter
2. bar-         bar graph of single para (optional cat_data)
3.hist-         hostogram of a bunch of data i.e. frequency of single para
4.scatter-      dot graph betweem two para
5.boxplot/box-  statical overview in form of box for a single para mean,outliers,quantiles(25%,75%),IQR (optional-catagorical data)
6.stem-         more ike bar graph,Lines instead of bars
7.violinplot-   violin shaped statistical overview graph spread of data(more like histogram)single para(optionl cat_data)

```python



x = np.linspace(0,5,11)
x1=np.arange(1,11,13)
```

```python
x
```

```python
x1
```

```python
y=x**3
```

```python
y2=x**4
```

```python
plt.plot(x,y)
```

```python

plt.bar(x,y)
plt.title('Bar Graph')
plt.show()

plt.hist(x,y)
plt.title('Histogram')
plt.show()

plt.scatter(x,y)
plt.title('Scatter')
plt.xlabel('x -- >')
plt.ylabel('y -->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()

plt.box(x,y)
plt.title('Boxplot')
plt.show()

```

```python



fig=plt.figure(figsize=(10,5))
plt.plot(x,y,lw=3,marker='o',markerfacecolor='red')
plt.plot(x,y2,marker='v',markersize=10)
plt.grid()
plt.legend(['y=x**3','y2=x**4'])
plt.show()
```
